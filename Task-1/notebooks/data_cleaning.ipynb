```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])
```

```python
df['Total_Cost'] = df['Unit_Cost'] * df['Quantity_Sold']
df['Profit'] = df['Sales_Amount'] - df['Total_Cost']
df['Year'] = df['Sale_Date'].dt.year
df['Month'] = df['Sale_Date'].dt.month

df.columns
```

```python
df['Region'] = df['Region'].str.strip().str.lower()
df['Sales_Rep'] = df['Sales_Rep'].str.strip()
```

```python
df.to_csv('../data/processed/cleaned_data.csv', index=False)
```

```python
df.head()
```

```python
#Sales by region
import matplotlib.pyplot as plt

df.groupby('Region')['Sales_Amount'].sum().plot(kind='bar')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
#Profit by month
df.groupby('Month')['Profit'].sum().plot(kind='line', marker='o')
plt.title('Monthly Profit Trend')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

## ✅ Final Summary

- Cleaned and prepared sales dataset for analysis
- No missing values or duplicates found
- Converted date column to datetime format
- Created new features: Total Cost, Profit, Year, Month
- Standardized categorical columns
