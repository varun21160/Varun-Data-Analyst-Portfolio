```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned_data.csv')
df.head()
```

##  Key KPIs

- Total Sales
- Total Profit
- Profit Margin
- Average Order Value
- Monthly Sales Growth

```python
total_sales = df['Sales_Amount'].sum()
total_profit = df['Profit'].sum()
profit_margin = total_profit / total_sales
avg_order_value = total_sales / len(df)

total_sales, total_profit, profit_margin, avg_order_value
```

## Profit by Category

```python
df.groupby('Product_Category')['Profit'].sum().plot(kind='bar')
plt.title('Profit by Category')
plt.show()
```

# Profit by Region

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title('Profit by Region')
plt.show()
```

# Profit Margin

```python
df['Profit_Margin'] = df['Profit'] / df['Sales_Amount']

df['Profit_Margin'].describe()
```

# Cost vs Sales

```python
plt.scatter(df['Total_Cost'], df['Sales_Amount'])
plt.xlabel('Total Cost')
plt.ylabel('Sales Amount')
plt.title('Cost vs Sales')
plt.show()
```

### Deep-Dive Insight:

The analysis reveals that the business is experiencing consistent losses due to high costs exceeding sales revenue across most transactions.

The negative average profit margin confirms that the majority of sales are not profitable.

The cost vs sales analysis further shows that many transactions have disproportionately high costs compared to revenue.

This indicates a critical issue in pricing strategy, cost control, or product profitability.

To improve performance, the business should focus on reducing costs, optimizing pricing, or identifying loss-making products.
